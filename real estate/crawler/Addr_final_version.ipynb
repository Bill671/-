```python
## 地址轉成經緯度
# https://smlpoints.com/guide-crawler-address-change-address-to-lantitude-and-longitude.html

# 改了side package的 source code : C:\Program Files(x86) 更正為 C:\Program Files (x86) 解決找不到瀏覽器問題
# https://stackoverflow.com/questions/71223000/could-not-get-version-for-google-chrome-with-the-command-powershell-erroracti

# https://towardsdatascience.com/%EF%B8%8F-load-the-same-csv-file-10x-times-faster-and-with-10x-less-memory-%EF%B8%8F-e93b485086c7

import os
import pandas as pd
import gc # 記憶體回收 https://stackify.com/python-garbage-collection/
import csv

def main(df,path,t):
    # import psutil # check mem info
    # proc = psutil.Process(os.getpid())
    # mem0 = proc.memory_info().rss

    from package.distance import get_coordinate
    import time
    import random
    
    # 逐筆 write 
    # By using with you close the file automatically and releases the memory.
    # https://stackoverflow.com/questions/39025431/how-can-i-release-memory-after-using-csv-writer-in-python
    # ref: https://blog.gtwang.org/programming/python-csv-file-reading-and-writing-tutorial/
    with open(path,'w', newline='') as csvfile:
        writer = csv.writer(csvfile)   
        writer.writerow(['Index', 'address','0']) # 寫入一列資料
        gc.collect() # release mem
        for col in df: # test df.head(2)
            i = 3
            while i>0:
                try: 
                    time.sleep(8 + random.uniform(0,10 ))
    #                 print(t,col)
                    writer.writerow([t,col,get_coordinate(col)])
                    i-=3
                    t+=1 # record index
                    gc.collect() # release mem
                except:
                    print(t,'err')
                    time.sleep(60)
                    i-=2
            if i==-1:
                t+=1 # record index
                writer.writerow([t, col,"nan"])  
                gc.collect()

            if t%2000 == 0: # 每2000筆印一筆資料
                print(t,'ok')

    # see the mem usage
    # pd = lambda x2, x1: 100.0 * (x2 - x1) / mem0  
    # mem1 = proc.memory_info().rss
    # print("Collect: %0.2f%%" % pd(mem1, mem0))
```
